```python
import numpy as np
import scipy
from scipy import integrate
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

### This notebook is the same as the submitted 2A but uses a different python solver that should more accurately represent ODE45. Maybe the solver I was using is why I didn't get the answer. The graphs look right to me. 

```python
def van_der_pol(t, y, e):
    #define system
    dv = y[1]
    dw = -e*(y[0]**2 - 1)*y[1]-y[0]
    return np.array([dv,dw])
```

```python
#define time and initial conditions
t0, t1 = 0, 32                    # start and end
t = np.linspace(t0, t1, 65)      # the points of evaluation of solution
y0_1 = [np.sqrt(3), 1]                       # initial value
y1 = np.zeros((len(t), len(y0_1)))   # array for solution
y1[0, :] = y0_1
```

```python
#define fitzhugh_1 inputs
e=0.1

#set up f1
r1 = integrate.ode(van_der_pol).set_integrator('dopri5')  # set integrator similiar to ODE45
r1.set_initial_value(y0_1, t0).set_f_params(e)   # initial values

#integrator loop
for i in range(1, t.size):
    #r1.set_f_params(e)
    y1[i, :] = r1.integrate(t[i])  # get one more value, add it to the array
    if not r1.successful():
        raise RuntimeError("Could not integrate")

p_1a = y1[:,0:1]
v_1a = y1[:,1:2]
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(t,p_1a, label = 'y1')
ax.plot(t,v_1a, label = 'y2')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('e=0.1')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
