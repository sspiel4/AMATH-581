```python
from __future__ import division

import numpy as np

import scipy
from scipy import integrate

from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
#online example
def fun(x, y):
    return np.vstack((y[1], -np.exp(y[0])))

#online example
def bc(ya, yb):
    return np.array([ya[0], yb[0]])

#online example
x = np.linspace(0,1,50)

y_a = np.zeros((2,len(x)))
y_b = np.zeros((2,len(x)))
y_b[0] = 3

res_a = integrate.solve_bvp(fun, bc, x, y_a)
res_b = integrate.solve_bvp(fun, bc, x, y_b)

y1a = res_a.sol(x)[0]
y1b = res_b.sol(x)[0]
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1a, label = '')
ax.plot(x, y1b, label = '')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('y')
ax.set_title('RHS')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

```python
#online example 2
def fun(x, y, p):
    k = p[0]
    return np.vstack((y[1], -k**2 * y[0]))

def bc(ya, yb, p):
    k = p[0]
    return np.array([ya[0], yb[0], ya[1]-k])

x = np.linspace(0,1,5)
y = np.zeros((2, len(x)))

y[0, 1] = 1
y[0, 3] = -1

sol = integrate.solve_bvp(fun, bc, x, y, p=[6])
```

```python
x_plot = np.linspace(0, 1, 100)
y_plot = sol.sol(x_plot)[0]
y2_plot = sol.sol(x_plot)[1]


#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(x_plot, y_plot, label = '')
ax.plot(x_plot, y2_plot, label = '')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('y')
ax.set_title('RHS')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
